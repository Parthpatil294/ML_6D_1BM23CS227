## Preprocess Dataset

### Subtask:
Handle missing values in the 'total_bedrooms' column by filling them with the column's median value.


**Reasoning**:
The subtask requires handling missing values in the 'total_bedrooms' column by filling them with the median value and then verifying the operation.



```python
import pandas as pd

# 1. Calculate the median value of the 'total_bedrooms' column
median_bedrooms = df['total_bedrooms'].median()
print(f"Median of 'total_bedrooms': {median_bedrooms}")

# 2. Replace missing values (NaNs) in 'total_bedrooms' with the median value
df['total_bedrooms'] = df['total_bedrooms'].fillna(median_bedrooms)

# 3. Verify the success of the operation
missing_count = df['total_bedrooms'].isnull().sum()
print(f"Number of missing values in 'total_bedrooms' after filling: {missing_count}")
```
